```python
# see if there is any mising data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Which brand has had the most Super Bowl commercials? Do they have a distinct style?

# to find out which brand had the most SB commericals
df['Brand'].value_counts()
```

```python
# Grabbing Just Bud Light

bl = df.loc[df['Brand']== 'Bud Light']

bl
```

```python
correlation_matrix = bl.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Features of Bud Light Commercials')

plt.xlabel('Commerical Features')
plt.ylabel('Commerical Features')


plt.show()


    
```
